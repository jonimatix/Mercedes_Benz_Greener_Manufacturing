```python
import category_encoders  as ce
from sklearn import decomposition, random_projection, manifold
import pandas as pd
```

## Dimension Reduction

```python
dt_encode = pd.read_csv("../../data/Mercedes_Benz_Greener_Manufacturing/data/R/dt_encode.csv")
```

```python
dt_encode.head()
```

```python
def featEng_dimRed(method, dt, n_component, cols):
        
    if method == "PCA":
        dr = decomposition.PCA(n_components = n_component, random_state = 420)
    elif method == "ICA":
        dr = decomposition.FastICA(n_components = n_component, random_state = 420)
    elif method == "SVD":
        dr = decomposition.TruncatedSVD(n_components = n_component, random_state = 420)
    elif method == "FA":
        dr = decomposition.FactorAnalysis(n_components = n_component, random_state = 420)
    elif method == "TSNE":
        dr = manifold.TSNE(n_components = n_component, random_state = 420)
    elif method == "GRP":
        dr = random_projection.GaussianRandomProjection(n_components = n_component, eps = 0.1, random_state = 420)
    elif method == "SRP":
        dr = random_projection.SparseRandomProjection(n_components = n_component, dense_output = True, random_state = 420)
        
    mx = dr.fit_transform(dt[cols])
    dt = dt.join(pd.DataFrame(mx, columns = [method + "_" + str(i) for i in range(1, n_component + 1)]))
    
    return dt
```

```python
cols_cat = ['X0', 'X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X8']
cols = dt_encode.drop(cols_cat + list(["y"]), axis = 1).columns.values
```

```python
methods_dr = ["PCA", "ICA", "SVD", "FA", "GRP", "SRP"] # TSNE
for method in methods_dr:
    dt_encode = featEng_dimRed(method, dt_encode, 12, cols)
```

```python
dt_encode.to_csv("../../data/Mercedes_Benz_Greener_Manufacturing/data/R/dt_dimensionReducce_fromPython.csv", index = False)
```
